```python
# in similay way as above we are transforming our doc 
# read lines int to words

docs = [l.split() for l in lines]
```

```python
print("Number of docs %d"%len(docs))
```

```python
# get the number of words 
print("Number of words : %d"% np.sum([len(d) for d in docs]))
```

```python
# now lets create the frequency vector for each word in the corpus
words = {}
for d in docs:
    for w in d:
        if(w not in words):
            words[w] = 1
        else:
            words[w] +=1
print("No of uniqu words : %d"%len(words))
# as the unique words will be equal to the keys of words obj
```

```python
# creating a mathplot of corpus word frequency
plt.plot(words.values())
```

```python
# creating a mathplot of corpus word frequency
wordValues = words.values()
plt.plot(sorted(wordValues))
plt.xlabel('word')
plt.ylabel('Log(Frequency)')
plt.yscale('log')
```

```python
from collections import defaultdict
df = defaultdict(int)
for d in docs:
    for w in set(d):
        df[w]+=1
plt.plot(wordValues)
# plt.yscale('log')
plt.xlabel('word')
plt.ylabel('frequency')
plt.title('Freqeuncy for words')
```
